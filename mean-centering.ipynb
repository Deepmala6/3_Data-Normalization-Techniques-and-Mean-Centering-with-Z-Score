```python
# # This libary use for mean - clustring.
import pandas as pd 
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("C:\\Users\\deepm\\Downloads\\100_Sales.csv")
df.head()
```

```python
# Drop specific columns
df_cleaned = df.drop(columns=["Unnamed: 9", "Unnamed: 10"], errors="ignore") # we are drop column.


df_cleaned.head()
```

```python
df_cleaned.describe()
```

```python
df.shape
```

```python
# Apply Z-score normalization on 'Total_Revenue' column
scaler = StandardScaler()
df_cleaned['Total_Revenue_Zscore'] = scaler.fit_transform(df_cleaned[['Total_Revenue']])

# Perform mean-centering on 'Total_Revenue' column
df_cleaned['Total_Revenue_Mean_Centered'] = df_cleaned['Total_Revenue'] - df_cleaned['Total_Revenue'].mean()

# Display the updated columns
print(df_cleaned[['Total_Revenue', 'Total_Revenue_Zscore', 'Total_Revenue_Mean_Centered']])
```
